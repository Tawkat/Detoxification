```python
import pandas as pd
import os

from sklearn.utils import shuffle

file_list = os.listdir('Generated_Data/Pred_3K_DataFrame')

columns = ['source_label', 'platform', 'source_file',
       'data_id', 'template', 'toxic_count', 'non_toxic_count', 'HateCheck',
       'stormfront', 'twitter', 'fb_yt', 'convAI', 'wikipedia',]

for file in file_list:
    if(not file.endswith('.csv')):
        continue
        
    df_new = {}
    df_new['toxic'] = []
    df_new['non_toxic'] = []
    for col in columns:
        df_new[col] = []
    df_ori = pd.read_csv(f'Generated_Data/Pred_3K_DataFrame/{file}') #.head() ######
    
    for ind in range(len(df_ori['source_label'])):
        if(df_ori['source_label'][ind]=='toxic'):
            df_new['toxic'].append(df_ori['source_text'][ind])
            df_new['non_toxic'].append(df_ori['target_text'][ind])
        elif(df_ori['source_label'][ind]=='non-toxic'):
            df_new['toxic'].append(df_ori['target_text'][ind])
            df_new['non_toxic'].append(df_ori['source_text'][ind])
        for col in columns:
            df_new[col].append(df_ori[col][ind])
        
    
    
    
    df_new = pd.DataFrame(df_new)

    for i in range(5):
        df_new = shuffle(df_new)
        df_new = df_new.reset_index(inplace=False, drop=True)

    df_new.to_csv(f'Datasets/Coverted_Col_Name_3K/{file}', index=False)
    
```

```python
import pandas as pd
import os
from tqdm import tqdm

file_list_1k = os.listdir('Generated_Data/Pred_DataFrame')

for file in tqdm(file_list_1k):
    if(not file.endswith('.csv')):
        continue
    platform = file.split('_1000')[0]
    if(platform=='facebook'):
        continue
    df_1k = pd.read_csv(f"Generated_Data/Pred_DataFrame/{file}")
    try:
        df_2k = pd.read_csv(f"NEW_Generated_Data/Pred_DataFrame/{platform}_2000.csv")
        df_3k = pd.concat([df_1k,df_2k], ignore_index=True)
        df_3k.to_csv(f"Generated_Data/Pred_3K_DataFrame/{platform}_3000.csv", index=False)
    except:
        df_1k.to_csv(f"Generated_Data/Pred_3K_DataFrame/{platform}_3000.csv", index=False)
        
        
```
